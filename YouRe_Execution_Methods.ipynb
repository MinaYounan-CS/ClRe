## Linear Execution Functions

```python
# Sequential Order
#========================================================================================================
def Linear_YouRe_30(lset):
    X=lset[0] 
    for i in range (1,len(lset)):
        Y=lset[i]
        newds= Alg.YouRe_30(X,Y)
        X=newds
    return newds

def Linear_YouRe_31(lset):
    X=lset[0] 
    for i in range (1,len(lset)):
        Y=lset[i]
        newds= Alg.YouRe_31(X,Y)
        X=newds
    return newds

def Linear_YouRe_32(lset):
    X=lset[0] 
    for i in range (1,len(lset)):
        Y=lset[i]
        newds= Alg.YouRe_32(X,Y)
        X=newds
    return newds
```

## Pair-merge Execution Functions

```python
def PairMerge_YouRe_30(lset):
    while len(lset)>1:
        odd=len(lset)%2
        NewList=[]
        count=0
        for i in range (0,len(lset)-odd,2):
            X=lset[i] 
            Y=lset[i+1]
            NewList.append(Alg.YouRe_30(X,Y))
        if(odd==1):
            X=lset[len(lset)-1]
            Y=NewList[len(NewList)-1]
            NewList[len(NewList)-1]=Alg.YouRe_30(X,Y)
        lset=NewList
    return lset[0]
#-----------------------------------------------------------------
def PairMerge_YouRe_31(lset):
    while len(lset)>1:
        odd=len(lset)%2
        NewList=[]
        count=0
        for i in range (0,len(lset)-odd,2):
            X=lset[i] 
            Y=lset[i+1]
            NewList.append(Alg.YouRe_31(X,Y))
        if(odd==1):
            X=lset[len(lset)-1]
            Y=NewList[len(NewList)-1]
            NewList[len(NewList)-1]=Alg.YouRe_31(X,Y)
        lset=NewList
    return lset[0]
#-----------------------------------------------------------------
def PairMerge_YouRe_32(lset):
    while len(lset)>1:
        odd=len(lset)%2
        NewList=[]
        count=0
        for i in range (0,len(lset)-odd,2):
            X=lset[i] 
            Y=lset[i+1]
            NewList.append(Alg.YouRe_32(X,Y))
        if(odd==1):
            X=lset[len(lset)-1]
            Y=NewList[len(NewList)-1]
            NewList[len(NewList)-1]=Alg.YouRe_32(X,Y)
        lset=NewList
    return lset[0]
```
